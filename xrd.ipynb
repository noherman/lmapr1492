```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
#Téléchargement de la structure
with MPRester("CEEbAW2taTmupi7NIXR") as m:
    structure = m.get_structure_by_material_id("mp-31451")
```

# Diffractogramme pour un rayonnement X incident Cu Kα  de longueur d'onde λ=1.54060 Å

```python
#Diffractogramme création & affichage
xrd = XRDCalculator('CuKa',1.54060)
pattern = xrd.get_pattern(structure, two_theta_range=(0,45))
xrd.show_plot(structure,annotate_peaks='compact', two_theta_range=(0,45))
```

# Résultats détaillés pour les 3 premiers pics

```python
for i in range(0,3):
    theta = pattern.x[i]
    intensity = pattern.y[i]
    hkl = pattern.hkls[i][0].get("hkl")
    multiplicity = pattern.hkls[i][0].get("multiplicity")
    if(i == 2):
        intensity = int(intensity)
    
    #Affichage des résultats
    print("Pour le pic numéro " + str(i+1) +": \n")
    print("- L'angle 2θ = " + str(theta) +" °\n")
    print("- L'intensité à ce pic vaut " +str(intensity) +" sur une échelle de 100 \n")
    print("- Les indices hkl qui lui sont associés sont " + str(hkl) +" et ont comme multiplicité : " + str(multiplicity) +"\n")
    print("\n")
```
